```python
import pandas as pd

qna_list = pd.read_excel('qna_total_list_210702.xlsx')
qna_list.head()

# qna_list.isna().sum()

# qna_list = qna_list.dropna()

# qna_list.isna().sum()

qna_list['q_length'] = qna_list['q_content'].str.len()

qna_list
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(qna_list['q_length'])
```

```python
import re
from konlpy.tag import Twitter
twt = Twitter()

def remove_html(sentence) :
    sentence = re.sub('(<([^>]+)>)', '', sentence)
    return sentence




stop_words = ['']
parts = ['Noun']
all_words = []
q_type = []
for n in range(len(qna_list)) : 
    text = str(qna_list['q_title'].iloc[n]) + str(qna_list['q_content'].iloc[n])
    text = remove_html(text.replace('_x000D_', ''))
#     print('ㅡ' * 30)
#     print(text)
    words = twt.pos(text)
    words_arr = []
    for i in words : 
        if not (i == 'EOS' or i == '') : 
            word_tmp = i[0]
            part = i[1]
            if part in parts : 
                if not(word_tmp in stop_words) : 
                    words_arr.append(word_tmp)
                    q_type.append(qna_list['q_type'].iloc[n])
    all_words.extend(words_arr)
print(all_words)
```

```python
all_words_df = pd.DataFrame({'words':all_words, 'q_type': q_type, 'count':len(all_words)*[1]})
words_q_type = all_words_df.groupby('words').mean()['q_type']
words_count = all_words_df.groupby('words').sum()['count']
words_df = pd.concat([words_q_type, words_count], axis=1)
# words_df.head()

words_df = words_df.loc[words_df['count'] >= 3]
words_df.sort_values('q_type', ascending=False).head()
```

```python
parts = ['Noun']
all_words_df = pd.DataFrame()
q_type = []
for n in range(len('qna_list')) : 
    text = str(qna_list['q_title'].iloc[n]) + str(qna_list['q_content'].iloc[n])
    text = remove_html(text.replace('_x000D_', ''))
    words = twt.pos(text)
    words_df = pd.DataFrame()
    for i in words : 
        if not (i == 'EOS' or i == '') : 
            word_tmp = i[0]
            part = i[1]
            if part in parts :
                words_df[word_tmp] = [1]
    all_words_df = pd.concat([all_words_df, words_df], ignore_index=True)
all_words_df.head()
```

```python
all_words_df = all_words_df.fillna(0)
all_words_df.head()
```

```python
content = str(qna_list['q_title'].iloc[2]) + str(qna_list['q_content'].iloc[2])
content = remove_html(content).replace('_x000D_', '')
# print(content)
target_text = all_words_df.iloc[2]
# print(target_text)
print(all_words_df)
```

```python
import numpy as np
cos_sim = []
for i in range(len(all_words_df)) :
    cos_text = all_words_df.iloc[i]
    cos = np.dot(target_text, cos_text) / (np.linalg.norm(target_text) * np.linalg.norm(cos_text))
    cos_sim.append(cos)
all_words_df['cos_sim'] = cos_sim
all_words_df.sort_values('cos_sim', ascending=False).head()
```
